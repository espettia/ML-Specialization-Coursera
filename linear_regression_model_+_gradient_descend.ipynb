```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x_train = np.array([1,2,3,4,5])
y_train = np.array([234,3,272,1500,1000])

#Tengo que ajustar los PARÁMETROS de un modelo
#El modelo a usar será "REGRESIÓN LINEAL"
#SE realiza a través de una función de error
def compute_prediction(w,b,x):
    return w*x+b

def sq_error_cost(w,b,x,y):
    total_cost=0
    m = len(x_train)
    for xi,yi in zip(x,y):
        total_cost = total_cost + (compute_prediction(w,b,xi)-yi)**2
    avg_cost = 1/(2*m)*total_cost
    return avg_cost

def partial_w(w,b,x,y):
    total_cost=0
    m = len(x_train)
    for xi,yi in zip(x,y):
        total_cost = total_cost + xi*(compute_prediction(w,b,xi)-yi)
    avg_cost = 1/(m)*total_cost
    return avg_cost

def partial_b(w,b,x,y):
    total_cost=0
    m = len(x_train)
    for xi,yi in zip(x,y):
        total_cost = total_cost + (compute_prediction(w,b,xi)-yi)
    avg_cost = 1/(m)*total_cost
    return avg_cost

def gradient_descent_linear_regression(x,y,w,b,a,n):
    for i in range(n):
        
        plt.plot(x_train,y_train,'o')
        plt.plot([0,max(x_train)],[b,compute_prediction(w,b,max(x_train))])
        plt.show()
        plt.clf()
        new_w = w - a*partial_w(w,b,x,y)
        new_b = b - a*partial_b(w,b,x,y)
        w=new_w
        b=new_b

    return [w,b]

w,b=gradient_descent_linear_regression(x_train,y_train,0,0,0.1,500)

plt.plot(x_train,y_train,'o')
plt.plot([0,max(x_train)],[b,compute_prediction(w,b,max(x_train))])
plt.show()
```
