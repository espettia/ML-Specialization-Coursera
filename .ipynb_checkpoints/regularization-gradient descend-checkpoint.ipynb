```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

X_train = np.arange(30,50)
X_train=X_train.reshape(-1,1)
Y_target = np.sin(X_train/2)
```

```python
plt.scatter(X_train,Y_target)
plt.title("Training data")
plt.show()
```

```python
X_eng = np.c_[X_train,X_train**2,X_train**3,X_train**4,X_train**5,X_train**6,X_train**7,X_train**8,X_train**9,X_train**10]
print(np.shape(X_eng))
```

```python
X_z = (X_eng-np.mean(X_eng,axis=0))/np.std(X_eng,axis=0)
```

```python
#w is a 2-D array
def dJ(X,Y,w,b):
    terms = X@w + b - Y
    partial = terms*np.c_[X,np.ones([len(X),1])]
    return np.sum(partial, axis=0).reshape(-1,1)

def GD(X,Y,w_ini,b_ini, it, alpha, lamb):
    pars = np.append(w_ini,b_ini)
    #I don't know if this works like this or if i should write it as having an indefinite number of parameters
    pars = pars.reshape(-1,1)
    for i in range(it):
        pars = pars*(1-alpha*lamb/len(X))-alpha*dJ(X,Y,pars[:-1],pars[-1])
    return pars

pars = np.random.rand(X_z.shape[1]+1)+5
pars_fin = GD(X_z,Y_target, pars[:-1],pars[-1],100000,0.01,100)
w=pars_fin[:-1]
b=pars_fin[-1]
Y_predict = X_z@w + b
plt.scatter(X_z[:,0],Y_target,label='Target')
plt.scatter(X_z[:,0],Y_predict,color='orange',label='Predict')
plt.legend()
plt.show()
```
